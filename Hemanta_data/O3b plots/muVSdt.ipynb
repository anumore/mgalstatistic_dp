```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
%matplotlib inline  
```

# Image type I
* data generated with the following condition
* strong lensed
* two image detectable
* both are type I
* the first and second column of lenstronomy output (data) 

```python
# getting data
f = open ('./detectable_4_image_type_1/lensed_params_detectable.json', "r")
param1 = json.loads(f.read())
```

```python
magnifications = np.array(param1['magnifications'])
magnifications[:,0]
```

```python
amp = param1['magnifications']
# magnification
amp = np.array(param1['magnifications'])
amp1_1 = amp[:,0]
amp1_2 = amp[:,1]
dmu12 = abs(amp1_2/amp1_1)

# time delays
dt = np.array(param1['time_delays'])/(60*60*24) #in days
delay1_1 = dt[:,0]
delay1_2 = dt[:,1]
dt12 = delay1_2-delay1_1

weights = np.array(param1['weights'])
```

```python
df = sns.load_dataset('iris')
# set seaborn style
sns.set_style("white")

# Basic 2D density plot
x = np.log10(dt12)
y = np.log(dmu12)
sns.kdeplot(x=x, y=y,legend=True, weights=weights, levels=1,thresh=.68, color='b') # 1 sigma (68%)
sns.kdeplot(x=x, y=y,legend=True, weights=weights, levels=1,thresh=.95, color='b') # 2 sigma (95%
sns.kdeplot(x=x, y=y,legend=True, weights=weights, levels=1,thresh=.997, color='b') # 3 sigma (99.7%)
plt.xlim(-2, 3)
plt.ylim(-2, 2)
plt.xlabel(r'Log (Time delay) (days)')
plt.ylabel(r'Log (Relative magnification)')
plt.title('relative time delay vs relative magnification \n Image type I')
plt.legend()
plt.grid()
plt.show()
```

# Image type II
* data generated with the following condition
* strong lensed
* two image detectable
* both are type II
* the third and fourth column of lenstronomy output (data) 

```python
# getting data
f = open ('./detectable_4_image_type_2/lensed_params_detectable.json', "r")
param2 = json.loads(f.read())
```

```python
amp = param2['magnifications']
# magnification
amp = np.array(param2['magnifications'])
amp2_3 = amp[:,2]
amp2_4 = amp[:,3]
dmu34 = abs(amp2_4/amp2_3)

# time delays
dt = np.array(param2['time_delays'])/(60*60*24)
delay2_3 = dt[:,2]
delay2_4 = dt[:,3]
dt34 = abs(delay2_3-delay2_4)
```

```python
df = sns.load_dataset('iris')
 
# set seaborn style
sns.set_style("white")

# Basic 2D density plot
x = np.log10(dt34)
y = np.log(dmu34)
sns.kdeplot(x=x, y=y,legend=True, levels=1,thresh=.68, color='b') # 1 sigma (68%)
sns.kdeplot(x=x, y=y,legend=True, levels=1,thresh=.95, color='b') # 2 sigma (95%
sns.kdeplot(x=x, y=y,legend=True, levels=1,thresh=.997, color='b') # 3 sigma (99.7%)
plt.xlim(-2, 3)
plt.ylim(-2, 2)
plt.xlabel(r'Log (Time delay) (days)')
plt.ylabel(r'Log (Relative magnification)')
plt.title('relative time delay vs relative magnification \n Image type II')
plt.legend()
plt.grid()
plt.show()
```
